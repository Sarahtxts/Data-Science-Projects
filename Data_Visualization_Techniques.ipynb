```python
import matplotlib.pyplot as plt
import seaborn as sns

months = range(1, 13)
actual = [120, 135, 150, 160, 170, 165, 180, 190, 200, 210, 205, 220]
predicted = [115, 140, 145, 155, 175, 160, 185, 195, 198, 208, 202, 225]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# Line plot
axes[0].plot(months, actual, label="Actual")
axes[0].plot(months, predicted, label="Predicted")
axes[0].set_title("Sales Trend")
axes[0].legend()

# Scatter plot
axes[1].scatter(actual, predicted, color="red")
axes[1].set_title("Actual vs Predicted")

# Histogram
axes[2].hist(actual, bins=5, color="skyblue", edgecolor="black")
axes[2].set_title("Sales Distribution")

sns.histplot(actual, bins=5, color="skyblue", edgecolor="black", kde=True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

exam1 = [45, 55, 60, 65, 70, 72, 74, 76, 78, 80,
         82, 84, 86, 88, 90, 92, 94, 96, 98, 100]

exam2 = [50, 58, 62, 68, 72, 74, 75, 78, 80, 83,
         85, 87, 88, 90, 91, 92, 95, 96, 97, 99]

# 1. Histogram of Exam 1 & Exam 2
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(exam1, bins=8, alpha=0.6, label="Exam 1")
plt.hist(exam2, bins=8, alpha=0.6, label="Exam 2")
plt.title("Score Distribution (Histogram)")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()

# 2. Density Plot
plt.subplot(1, 2, 2)
sns.kdeplot(exam1, shade='True',label="Exam 1")
sns.kdeplot(exam2, shade='True',label="Exam 2")
plt.title("Score Distribution (Density Plot)")
plt.xlabel("Scores")
plt.ylabel("Density")
plt.legend()

plt.tight_layout()
plt.show()

# 3. Scatter Plot
plt.figure(figsize=(6, 5))
plt.scatter(exam1, exam2, color="purple")
plt.title("Relationship Between Exam 1 and Exam 2 Scores")
plt.xlabel("Exam 1 Scores")
plt.ylabel("Exam 2 Scores")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

exam1 = [45, 55, 60, 65, 70, 72, 74, 76, 78, 80,
         82, 84, 86, 88, 90, 92, 94, 96, 98, 100]

# Mean of scores
mean_score = np.mean(exam1)

# Histogram
plt.hist(exam1, bins=8, color="skyblue", edgecolor="black")
plt.axvline(mean_score, color="red", linestyle="--", label=f"Mean = {mean_score:.1f}")

plt.title("Exam 1 Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset("iris")

# Scatter plot of sepal length vs sepal width
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
correlation_matrix = iris.drop("species", axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2000, 2001, 2002, 2003, 2004]
savings = [2000, 4000, 9000, 14000, 16000]

plt.plot(years, savings, marker='o')
plt.axvline(2002,color='green',linestyle='--')

# Annotating the point where the money goal is reached
plt.annotate(
    'My Money Goal Has been Reached!',
    xy=(2003, 14000),
    xytext=(2002, 20000),
    arrowprops=dict(facecolor='red')
)

plt.title('Savings Over Years')
plt.xlabel('Year')
plt.ylabel('Amount Saved')
plt.grid(True)
plt.show()
```

```python

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

ax = plt.axes(projection='3d')


z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# 3d plot
ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].plot(x, np.sin(x))
ax[0, 0].set_title('Sine')

ax[0, 1].plot(x, np.cos(x), 'r')
ax[0, 1].set_title('Cosine')

ax[1, 0].plot(x, np.tan(x), 'g')
ax[1, 0].set_ylim(-10, 10)
ax[1, 0].set_title('Tangent')

ax[1, 1].plot(x, np.exp(-x), 'm')
ax[1, 1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
        'Value': [10, 15, 12, 18, 20]}
df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])

# Create a time series plot
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Value'], marker='o', linestyle='-', color='green')
plt.title('Simple Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
